# Project Overview
This project demonstrates basic data analysis techniques using **Python**.  
We read a CSV file, perform data exploration, and visualize some features.


```python
# Required libraries import karna
import pandas as pd
import matplotlib.pyplot as plt

# Loading Dataset
df = pd.read_csv("Student_performance_data.csv")
print(df.head())
```

```python
# Shape of Dataset
print("\n Dataset Shape (rows, columns):", df.shape)
```

```python
# Summary Statistics
print("\nüîπ Summary Statistics:")
print(df.describe())
```

```python
# Missing Values
print("\nüîπ Missing Values:")
print(df.isnull().sum())
```

```python
if "Age" in df.columns:
    plt.hist(df['Age'], bins=10, color='skyblue', edgecolor='black')
    plt.title("Age Distribution")
    plt.xlabel("Age")
    plt.ylabel("Frequency")
    plt.show()
else:
    print("Age Column is not in Dataset.")
```

```python
# Syntax: df['column'].value_counts().plot(kind='bar')
if "Gender" in df.columns:
    df['Gender'].value_counts().plot(kind='bar', color=['lightgreen', 'lightcoral'])
    plt.title("Gender Count")
    plt.xlabel("Gender")
    plt.ylabel("Count")
    plt.show()
else:
    print("‚ö†Ô∏è Gender column dataset not found.")
```
